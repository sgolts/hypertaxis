```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sys
import os
import glob
```

```python
dpath = "/scratch/indikar_root/indikar1/shared_data/single_cell/align_table/"
file_list = glob.glob(f"{dpath}*")
# shuffle 'em'
np.random.shuffle(file_list)
file_list[:10]
```

```python
sample_size = 10

result = []

for i, file_path in enumerate(file_list[:sample_size]):

    basename = os.path.basename(file_path)
    sample_id = basename[:2]
    file_id = basename[2:5]
    df = pd.read_parquet(file_path)
    df['basename'] = basename
    df['sample_id'] = sample_id
    df['file_id'] = file_id
    result.append(df)
    
    
result = pd.concat(result)
result.head()
```

```python
random_read = result['read_name'].sample(1).values[0]

columns = [
    'read_name',
    'read_start',
    'read_end',
    'chrom',
    'fragment_id',
    'mapping_quality',
]

read_df = result[result['read_name'] == random_read]
read_df[columns]
```
