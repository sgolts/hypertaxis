```python
def generate_incidence_matrix(n_rows, n_cols, order, min_count):
    """Generates an incidence matrix with randomly imputed new columns.

    Args:
        n_rows: The number of rows in the incidence matrix.
        n_cols: The number of new columns to generate.
        order: The approximate mean number of 1s per column.
        min_count: The minimum number of 1s allowed in a generated column.

    Returns:
        A pandas DataFrame representing the imputed incidence matrix.
    """

    counts = np.random.poisson(order, size=n_cols).astype(int)
    counts = np.maximum(counts, min_count)  # Ensure minimum count

    columns = []
    for i in range(n_cols):
        col = np.zeros(n_rows, dtype=int)
        col[np.random.choice(n_rows, size=counts[i], replace=False)] = 1
        columns.append(col)

    return pd.DataFrame(columns).T 


n_rows = 10
n_cols = 10
I = generate_incidence_matrix(n_rows=n_rows, 
                              n_cols=n_cols, 
                              order=3, 
                              min_count=3)

I.head()
```

```python
def plot_network(G, node_kwargs={}, edge_kwargs={}, plot_nodes=True, plot_edges=True, labels=None, label_kwargs={}):
    """
    Plots a networkx graph with customizable options, control over plotting, and optional node labels.

    Args:
        G: A networkx.Graph object representing the network.
        node_kwargs: Keyword arguments for customizing nodes.
        edge_kwargs: Keyword arguments for customizing edges.
        plot_nodes: Boolean flag to control if nodes are plotted (default: True).
        plot_edges: Boolean flag to control if edges are plotted (default: True).
        labels:  A dictionary of node labels keyed by node (optional).
        label_kwargs: Additional keyword arguments to be passed to nx.draw_networkx_labels. 
    """

    if plot_nodes:
        node_options = {
            'node_size': 100,
            'node_color': 'grey',
            'edgecolors': 'k',
            'linewidths': 1
        }
        node_options.update(node_kwargs)  
        nx.draw_networkx_nodes(G, G.pos, **node_options)

    if plot_edges:
        edge_options = {
            'edge_color': 'k',
            'width': 3
        }
        edge_options.update(edge_kwargs)  
        nx.draw_networkx_edges(G, G.pos, **edge_options)

    if labels:
        nx.draw_networkx_labels(G, G.pos, labels, **label_kwargs)
        
        
def remove_edges(G, p=0.5):
    """
    Takes a graph and randomly removes edges with a given probability.

    Args:
        G: The input networkx.Graph.
        p: The probability of removing an edge.

    Returns:
        networkx.Graph: A copy of the graph with edges removed.
    """
    G_modified = G.copy()
    edges = np.array(G_modified.edges())

    for u, v in edges:
        if np.random.rand() < p:
            G_modified.remove_edge(u, v)
    
    G_modified.pos = G.pos
    return G_modified
        
# Plot the "core"
node_options = {
    'node_size': 15,
    'node_color': 'grey',
    'edgecolors': 'k',
    'linewidths': 0.5
}

edge_options = {
    'edge_color': 'k',
    'width': 0.75,
    'alpha' : 0.1,
}


G = nx.from_numpy_array(A.to_numpy())
G.pos = nx.spring_layout(G)
G = remove_edges(G, p=0.75)

plt.rcParams['figure.dpi'] = 200
plt.rcParams['figure.figsize'] = 5, 5
plot_network(G, 
             node_kwargs=node_options, 
             edge_kwargs=edge_options, 
            )
```
