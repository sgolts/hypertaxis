```python
import copy
import pandas as pd
import numpy as np
import networkx as nx
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
import random
import sys
from importlib import reload

sys.path.append("../")
import utils as ut
import plotting as plt2
```

```python
def generate_incidence_matrix(n_rows, n_cols, order, min_count):
    """Generates an incidence matrix with randomly imputed new columns.

    Args:
        n_rows: The number of rows in the incidence matrix.
        n_cols: The number of new columns to generate.
        order: The approximate mean number of 1s per column.
        min_count: The minimum number of 1s allowed in a generated column.

    Returns:
        A pandas DataFrame representing the imputed incidence matrix.
    """

    counts = np.random.poisson(order, size=n_cols).astype(int)
    counts = np.maximum(counts, min_count)  # Ensure minimum count

    columns = []
    for i in range(n_cols):
        col = np.zeros(n_rows, dtype=int)
        col[np.random.choice(n_rows, size=counts[i], replace=False)] = 1
        columns.append(col)

    return pd.DataFrame(columns).T 


n_rows = 193
n_cols = 150
I = generate_incidence_matrix(n_rows=n_rows, 
                              n_cols=n_cols, 
                              order=3, 
                              min_count=3)
I.shape
```

```python
reload(plt2)
p_red = 0.025
px = np.random.random(n_cols)
colors = list(np.where(px < p_red, 1.0, 0.0))
colors = plt2.floats_to_colors(colors, 
                               vmin=0.05, 
                               vmax=0.99, 
                               colormap='Set1_r')

node_params = {
    's' : 15,
    'ec' : 'k',
    'lw' : 0.25,
    'zorder' : 2,
}

line_params = {
    'lw' : 0.5,
    'alpha' : 0.5,
    'zorder' : 1,
}
    
plt.rcParams['figure.dpi'] = 200
plt.rcParams['figure.figsize'] = 12, 5
plt2.plot_incidence(ut.sort_by_lowest_index(I),
                    node_color=colors,
                    node_params=node_params,
                    line_params=line_params)

plt.yticks([])
plt.xticks([])
```
