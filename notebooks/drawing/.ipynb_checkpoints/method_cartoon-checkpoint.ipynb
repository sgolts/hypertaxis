```python
def generate_small_world_graph(n=25, k=5, p=0.1):
    """
    Generates a small-world graph with n nodes using the Watts-Strogatz model.

    Args:
        n: Number of nodes.
        k: Each node is connected to k nearest neighbors in a ring topology.
        p: Probability of rewiring each edge. 

    Returns:
        networkx.Graph: The generated small-world graph.
    """

    G = nx.watts_strogatz_graph(n, k, p)
    G.pos = nx.spring_layout(G)
    return G


def rewire_edges(G, p=0.3):
    """
    Takes a graph and rewires edges with a given probability.

    Args:
        G: The input networkx.Graph.
        p: The probability of rewiring an edge.

    Returns:
        networkx.Graph: A copy of the graph with rewired edges.
    """

    G_rewired = G.copy()
    edges = np.array(G_rewired.edges())

    for u, v in edges:
        if np.random.rand() < p:  # Check if rewiring should occur
            G_rewired.remove_edge(u, v)

            available_nodes = [n for n in G.nodes() if n != u and not G_rewired.has_edge(u, n)]
            if available_nodes:
                new_v = np.random.choice(available_nodes)
                G_rewired.add_edge(u, new_v)  

    G_rewired.pos = G.pos
    return G_rewired


def remove_edges(G, p=0.5):
    """
    Takes a graph and randomly removes edges with a given probability.

    Args:
        G: The input networkx.Graph.
        p: The probability of removing an edge.

    Returns:
        networkx.Graph: A copy of the graph with edges removed.
    """
    G_modified = G.copy()
    edges = np.array(G_modified.edges())

    for u, v in edges:
        if np.random.rand() < p:
            G_modified.remove_edge(u, v)
    
    G_modified.pos = G.pos
    return G_modified


def plot_network(G, node_kwargs={}, edge_kwargs={}, plot_nodes=True, plot_edges=True, labels=None, label_kwargs={}):
    """
    Plots a networkx graph with customizable options, control over plotting, and optional node labels.

    Args:
        G: A networkx.Graph object representing the network.
        node_kwargs: Keyword arguments for customizing nodes.
        edge_kwargs: Keyword arguments for customizing edges.
        plot_nodes: Boolean flag to control if nodes are plotted (default: True).
        plot_edges: Boolean flag to control if edges are plotted (default: True).
        labels:  A dictionary of node labels keyed by node (optional).
        label_kwargs: Additional keyword arguments to be passed to nx.draw_networkx_labels. 
    """

    if plot_nodes:
        node_options = {
            'node_size': 100,
            'node_color': 'grey',
            'edgecolors': 'k',
            'linewidths': 1
        }
        node_options.update(node_kwargs)  
        nx.draw_networkx_nodes(G, G.pos, **node_options)

    if plot_edges:
        edge_options = {
            'edge_color': 'k',
            'width': 3
        }
        edge_options.update(edge_kwargs)  
        nx.draw_networkx_edges(G, G.pos, **edge_options)

    if labels:
        nx.draw_networkx_labels(G, G.pos, labels, **label_kwargs)



plt.rcParams['figure.dpi'] = 200
plt.rcParams['figure.figsize'] = 5, 5


n = 35
labels = {x:x+1 for x in list(range(n))}
G_core = generate_small_world_graph(n=n, k=6, p=0.2)
G_core = remove_edges(G_core, p=0.25)

sample_size = 100
samples = []
for i in range(sample_size):
    G_sc = rewire_edges(G_core)
    G_sc = remove_edges(G_sc)
    G_sc.pos = G_core.pos
    samples.append(copy.deepcopy(G_sc))
    edge_options = {
        'edge_color': 'k',
        'width': 0.75,
        'alpha' : 0.1,
    }
    plot_network(G_sc, edge_kwargs=edge_options, plot_nodes=False)


# Plot the "core"
node_options = {
    'node_size': 100,
    'node_color': 'grey',
    'edgecolors': 'k',
    'linewidths': 1
}

edge_options = {
    'edge_color': 'red',
    'width': 3,
    'alpha' : 0.5,
}

plot_network(G_core, 
             node_kwargs=node_options, 
             edge_kwargs=edge_options, 
             # labels=labels,
             # label_kwargs={'font_size' : 5, "font_weight" : 'bold'}
            )
```

```python
def plot_polygons(G, ax=None, k=3, polygons=None):
    """
    Highlights polygons of degree k in an existing graph plot.

    Args:
        G: A networkx.Graph object (needed for node positions).
        ax: An existing matplotlib.axes.Axes object (optional).
        k: The degree of polygons to find and plot.
        polygons: A list of polygons, where each polygon is a list of nodes.
    """

    if ax is None:
        fig, ax = plt.subplots()

    if polygons is None:
        # Find polygons of degree k 
        polygons = [list(T) for T in nx.find_cliques(G) if len(T) == k]

    # Highlight polygons
    for i, polygon_nodes in enumerate(polygons):
        poly_coords = [G.pos[node] for node in polygon_nodes]
        polygon = Polygon(poly_coords, 
                          closed=True, 
                          alpha=0.75, 
                          edgecolor='r', 
                          facecolor='lightblue')
        ax.add_patch(polygon)
        

g = recon.create_graph_tools_from_adjacency(A)
hyperedges = recon.learn_hyperedges_mcmc(g, niter=100)

order_t = 3
hyperedges = [x for x in hyperedges if len(x) >= order_t]
len(hyperedges)
print(hyperedges)
        
plot_network(G_core, 
             node_kwargs=node_options, 
             edge_kwargs=edge_options, 
            )

plot_polygons(G_core, ax=plt.gca(), polygons=hyperedges)

plt.show()
```

```python
def impute_incidence_columns(incidence_df, num_new_columns, min_order):
    """
    Randomly imputes new columns into an incidence DataFrame, preserving column sum distribution.

    The number of 1s in each column is approximated by a Poisson distribution, controlled by 
    

    Args:
        incidence_df: A pandas DataFrame representing the incidence matrix.
        num_new_columns: The number of new columns to generate.

    Returns:
        A new pandas DataFrame with the imputed columns appended.
    """
    m, n = incidence_df.shape
    # Calculate column sum distribution of the input 
    column_sums = incidence_df.sum()
    order = np.mean(column_sums).astype(int)
    print(f"{order=}")
    cardinalities = np.random.poisson(order, size=num_new_columns).astype(int)
    print(f"{incidence_df.shape=} {cardinalities.shape=}")
    
    # handle low-order contacts
    cardinalities = np.where(cardinalities <= min_order, min_order, cardinalities)
    
    
    # Generate new columns matching the distribution
    new_columns = []
    for i in range(num_new_columns):
        column = np.zeros(m)
        hyperedge_ind = np.random.choice(column_sums.index, 
                                  cardinalities[i], 
                                  replace=False,)
        column[hyperedge_ind] = 1
        new_columns.append(column)

    new_columns_df = pd.DataFrame(new_columns).T

    # Combine input and new columns
    imputed_df = pd.concat([incidence_df, new_columns_df], axis=1, ignore_index=True)
    imputed_df = imputed_df.fillna(0)
    return imputed_df.astype(int)

I_pop = impute_incidence_columns(I, num_new_columns=30, min_order=3)
I_pop


node_params = {
    's' : 35,
    'ec' : 'k',
    'lw' : 0.75,
    'zorder' : 2,
}

line_params = {
    'lw' : 1,
    'zorder' : 1,
}

plt.rcParams['figure.dpi'] = 200
plt.rcParams['figure.figsize'] = 7, 5
plt2.plot_incidence(ut.sort_by_lowest_index(I_pop),
                    node_color='grey',
                    line_color='k',
                    node_params=node_params,
                    line_params=line_params)
plt.xticks([])
plt.yticks([])
```
