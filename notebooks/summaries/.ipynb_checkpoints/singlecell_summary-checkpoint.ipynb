# load the data

```python
# load ann data
resolution = 1000000
fpath = f"/scratch/indikar_root/indikar1/shared_data/higher_order/lightweight/singlecell_mESC_{resolution}_anndata.h5ad"

start_time = time.time()  # Record the start time
adata = sc.read_h5ad(fpath)
end_time = time.time()  # Record the end time
print(f"Time taken to read the file: {end_time - start_time:.2f} seconds")
sc.logging.print_memory_usage()

# sort the object by global bin id
sorted_indices = adata.obs['bin'].argsort()
adata = adata[sorted_indices, :] 

adata
```

```python
adata.uns['chrom_sizes']['n_bins'] = adata.uns['chrom_sizes']['size'] // adata.uns['base_resolution']
adata.uns['chrom_sizes'].head()
```

# Add the core scores

```python
fpath = "/scratch/indikar_root/indikar1/shared_data/higher_order/global_core_score/population_mESC_1000000_scores.csv"
df = pd.read_csv(fpath)

columns_to_drop = [x for x in df.columns if x in adata.obs.columns]
df = df.drop(columns=columns_to_drop)
print(f"{df.shape=}")

adata.obs = pd.merge(
    adata.obs, df.set_index('bin_name'),
    how='left',
    left_index=True,
    right_index=True,
)

adata.obs.head()
```

```python
# remove outlier bins from population
remove_bins = adata.obs[adata.obs['degree_outlier'].astype(bool)].index.to_list()
print(f"Removing top {len(remove_bins)} outlier loci: ")
print(remove_bins)

adata = adata[~adata.obs_names.isin(remove_bins), :].copy()

adata
```

# Load the duplicate reports



```python
dpath = f"/scratch/indikar_root/indikar1/shared_data/higher_order/duplicates/singlecell_mESC_{resolution}_chr*.parquet"
file_list = glob.glob(dpath)
print(f"n files: {len(file_list)}")

df = []

for fpath in file_list:
    chrom = os.path.basename(fpath).split("_")[3].split(".")[0]
    print(f"{chrom}...")
    tmp = pd.read_parquet(fpath)
    tmp['chrom'] = chrom
    df.append(tmp)

df = pd.concat(df)
df = df.sort_values(by=['read_id', 'mapping_quality'], ascending=[True, False])
df = df.drop_duplicates(subset=['read_id', 'cell_id'])

print(df['unique'].value_counts())
print(f"{df.shape=}")
df.head()
```

# drop non-unique reads

```python
order_threshold = 1 # at least pairwise

duplicate_reads = df[~df['unique']]['read_id'].values
print(f"{duplicate_reads.shape=}")

mask = ~adata.var_names.isin(duplicate_reads)
bdata = adata[:, mask].copy()
bdata = bdata[:, bdata.X.sum(axis=0) > order_threshold].copy() # drop singletons

bdata.var['new_order'] = np.ravel(bdata.X.sum(axis=0))
bdata.var['higher_order'] = bdata.var['new_order'] > 2

bdata
```

# Cell coverage

```python
H = bdata.to_df()
H.columns = bdata.var['basename'].values
print(f"{H.shape=}")

def get_nonzero_cols(row):
  """Returns a list of column names with nonzero values for a given row."""
  return len(set(H.columns[row.astype(bool)]))

cell_coverage = pd.DataFrame(H.apply(get_nonzero_cols, axis=1), columns=['n_cells'])

del H

cell_coverage['percent_cells'] = 100*(cell_coverage['n_cells'] / 380)
cell_coverage = cell_coverage.reset_index(drop=False)

cell_coverage['chrom'] = cell_coverage['bin_name'].str.split(":").str[0]
cell_coverage['chrom_bin'] = cell_coverage['bin_name'].str.split(":").str[1]

# merge features
columns = [
    'bin_name',
    'ATACSeq_1', 
    'CTCF',
    'H3K27ac', 
    'H3K27me3',
    'RNA_5',
    'ce_singular_vector_1', 
    'ce_eigenvector_centrality',
    'hge_singular_vector_1',
    'hge_logexp_RNA_weighted', 
    'global_singular_vector_1',
    'global_hge_logexp_RNA_weighted'
]

cell_coverage = pd.merge(
    cell_coverage,
    adata.obs.reset_index(drop=False)[columns],
    how='left',
    left_on='bin_name',
    right_on='bin_name',
)

cell_coverage.head()
```

```python
print(cell_coverage['n_cells'].mean())
```

```python
plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.figsize'] = 4, 2.5

sns.barplot(
    data=cell_coverage,
    x='chrom',
    y='percent_cells',
    edgecolor='k',
    lw=1,
    width=0.75,
    color='lightgrey',
)

plt.gca().tick_params(axis='x', rotation=90)
plt.ylabel('% single cells')
plt.xlabel('')
sns.despine()
```

```python
score_column = 'global_hge_logexp_RNA_weighted'
core_threshold_quantile = 0.75

pdf = cell_coverage.copy()
threshold = np.quantile(pdf[score_column], core_threshold_quantile)

print(f"Using score column: {score_column}")
print(f"Core threshold quantile: {core_threshold_quantile}")
print(f"Calculated threshold: {threshold:.3f}") 

pdf['is_core'] = np.where(pdf[score_column] > threshold, 'core', 'non-core')

# Calculate and print means for both groups
core_mean = pdf[pdf['is_core'] == 'core']['n_cells'].mean()
non_core_mean = pdf[pdf['is_core'] == 'non-core']['n_cells'].mean()

print(f"Mean percent_cells for core samples: {core_mean:.3f}")
print(f"Mean percent_cells for non-core samples: {non_core_mean:.3f}")

# Perform independent t-test
core_cells = pdf[pdf['is_core'] == 'core']['n_cells']
non_core_cells = pdf[pdf['is_core'] == 'non-core']['n_cells']
t_stat, p_value = scipy.stats.ttest_ind(core_cells, non_core_cells)

print(f"Independent t-test results:")
print(f"  t-statistic: {t_stat:.3f}")
print(f"  p-value: {p_value:.3f}")

plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.figsize'] = 1.2, 2.5

sns.boxplot(
    data=pdf,
    x='is_core',
    y='percent_cells',
     width=0.5,
    showfliers=False,
    linecolor='k',
    palette=['lightgrey', 'red'],
)

plt.gca().tick_params(axis='x', rotation=90)

plt.ylabel('% single cells')
plt.xlabel('')

sns.despine()
```

# Plot the incidence matrix

```python
chromosome = "1"
order_threshold = 2
n_cells = 379

cell_ids = bdata.var['basename'].unique()
cell_ids = np.random.choice(cell_ids, n_cells, replace=False)

# sample single-cells
sample = bdata[bdata.obs['chrom'] == chromosome, bdata.var['basename'].isin(cell_ids)].copy()
read_map = dict(zip(sample.var_names, sample.var['basename'].values))

H = sample.to_df()
H = H.T.drop_duplicates().T

H = H.loc[:, (H.sum() >= order_threshold)]
print(f"(raw) {H.shape=}")
H = ut.sort_by_lowest_index(H)

rows, cols = np.nonzero(H)

# Create a list to store the cell names corresponding to each point
cell_names = []
for row, col in zip(rows, cols):
    cell_name = read_map[H.columns[col]]
    cell_names.append(cell_name)

# Create a DataFrame with columns and rows
df_points = pd.DataFrame({
    'col': cols, 
    'row': rows,
    'cell': cell_names,
})

print(f"{df_points.shape=}")

plt.rcParams['figure.dpi'] = 200
plt.rcParams['figure.figsize'] = 10, 3.5

pdf = df_points.copy()
# the coordinates of the reference point
min_col = 0
max_row = 1

# Calculate the squared distances from (min_col, max_row)
pdf['distance'] = (ut.min_max(pdf['row']) - max_row)**2 + (ut.min_max(pdf['col']) - min_col)**2 

# Calculate the Euclidean distance
pdf['distance'] = np.sqrt(pdf['distance'])

sns.scatterplot(
    data=pdf,
    x='col',
    y='row',
    marker=".",
    size='distance',
    sizes=(0.1, 15),
    alpha=0.65,
    ec='none',
    color='k',
    legend=False,
)

plt.gca().invert_yaxis()
plt.ylabel(f'chromosome {chromosome}')
plt.xlabel('scpore-c read')
plt.xticks([])
sns.despine()

df_points.head()
```
