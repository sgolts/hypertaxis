```python
plt.rcParams['figure.dpi'] = 200
plt.rcParams['figure.figsize'] = (9, 4)

# Centralities
bc = nx.betweenness_centrality(G)          # left
ec = nx.eigenvector_centrality(G)          # right

cmap = cm.get_cmap("bwr")

fig, axes = plt.subplots(1, 2)

# --- Left: Betweenness ---
ax = axes[0]
vmin, vmax = min(bc.values()), max(bc.values())
norm = colors.Normalize(vmin=vmin, vmax=vmax)
node_colors = [cmap(norm(bc[n])) for n in G.nodes()]
nx.draw(
    G, G.pos, ax=ax,
    node_size=400, node_color=node_colors,
    edgecolors='black', linewidths=1.0,
    width=4, edge_color='grey',
    with_labels=True, font_size=10
)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
plt.colorbar(sm, ax=ax, shrink=0.7, label="Betweenness")
ax.set_title("Betweenness centrality")

# --- Right: Eigenvector ---
ax = axes[1]
vmin, vmax = min(ec.values()), max(ec.values())
norm = colors.Normalize(vmin=vmin, vmax=vmax)
node_colors = [cmap(norm(ec[n])) for n in G.nodes()]
nx.draw(
    G, G.pos, ax=ax,
    node_size=400, node_color=node_colors,
    edgecolors='black', linewidths=1.0,
    width=4, edge_color='grey',
    with_labels=True, font_size=10
)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
plt.colorbar(sm, ax=ax, shrink=0.7, label="Eigenvector")
ax.set_title("Eigenvector centrality")

fig.tight_layout()
fig.set_facecolor('none')
plt.show()
```

# Curvature

```python
# Two-panel ORC plots: edges (left), node-mean (right)

plt.rcParams['figure.dpi'] = 200
fig, axes = plt.subplots(1, 2, figsize=(9, 4))

# ---- Compute Ollivier–Ricci curvature once ----
orc = OllivierRicci(G, alpha=0.5)
orc.compute_ricci_curvature()

key = 'ricciCurvature'
edge_attr = nx.get_edge_attributes(orc.G, key)

# Use undirected-safe keys
edge_curv = {frozenset((u, v)): c for (u, v), c in edge_attr.items()}

# ---------------- Left: edges colored by ORC ----------------
ax = axes[0]
vals_edges = list(edge_curv.values())
vmin_e, vmax_e = min(vals_edges), max(vals_edges)
norm_e = colors.Normalize(vmin=vmin_e, vmax=vmax_e)
cmap_e = cm.get_cmap("coolwarm")

edge_colors = [cmap_e(norm_e(edge_curv[frozenset((u, v))])) for u, v in G.edges()]

nx.draw_networkx_nodes(G, G.pos, node_color="lightgray", edgecolors="black",
                       node_size=400, ax=ax)
nx.draw_networkx_edges(G, G.pos, edge_color=edge_colors, width=4, ax=ax)
nx.draw_networkx_labels(G, G.pos, font_size=10, ax=ax)

sm_e = plt.cm.ScalarMappable(cmap=cmap_e, norm=norm_e)
sm_e.set_array([])
plt.colorbar(sm_e, ax=ax, shrink=0.7, label="Ollivier–Ricci (edges)")
ax.set_title("Edge curvature")
ax.axis("off")

# ------------- Right: nodes colored by mean incident ORC -------------
ax = axes[1]
node_mean = {}
for n in G.nodes():
    inc = [edge_curv[frozenset((n, v))] for v in G.neighbors(n)]
    node_mean[n] = float(np.mean(inc)) if inc else 0.0

vals_nodes = list(node_mean.values())
vmin_n, vmax_n = min(vals_nodes), max(vals_nodes)
norm_n = colors.Normalize(vmin=vmin_n, vmax=vmax_n)
cmap_n = cm.get_cmap("bwr_r")

node_colors = [cmap_n(norm_n(node_mean[n])) for n in G.nodes()]

nx.draw_networkx_nodes(G, G.pos, node_color=node_colors, edgecolors="black",
                       node_size=400, ax=ax)
nx.draw_networkx_edges(G, G.pos, edge_color="grey", width=4, ax=ax)
nx.draw_networkx_labels(G, G.pos, font_size=10, ax=ax)

sm_n = plt.cm.ScalarMappable(cmap=cmap_n, norm=norm_n)
sm_n.set_array([])
plt.colorbar(sm_n, ax=ax, shrink=0.7, label="Node mean ORC")
ax.set_title("Node mean curvature")
ax.axis("off")

fig.tight_layout()
fig.set_facecolor('none')
plt.show()
```

```python
# --- Ricci curvature has already been computed on orc.G ---
key = 'ricciCurvature'
edge_curv = nx.get_edge_attributes(orc.G, key)

# Make lookup independent of edge order (undirected graphs)
edge_curv_u = {frozenset(e): v for e, v in edge_curv.items()}

# Node-wise mean of incident edge curvatures
node_ricci_mean = {}
for n in G.nodes():
    vals = [
        edge_curv_u[frozenset((n, nbr))]
        for nbr in G.neighbors(n)
        if frozenset((n, nbr)) in edge_curv_u
    ]
    node_ricci_mean[n] = float(np.mean(vals)) if vals else 0.0  # use 0.0 for isolated nodes

# Centralities
eig = nx.eigenvector_centrality(G, max_iter=1000)      # increase max_iter for safety
btw = nx.betweenness_centrality(G, normalized=True)

# Assemble DataFrame
df = (
    pd.DataFrame.from_dict(node_ricci_mean, orient='index', columns=['ricci_mean'])
      .join(pd.Series(eig, name='eigenvector'))
      .join(pd.Series(btw, name='betweenness'))
      .rename_axis('node')
)

print(df)
```

```python
# Compute correlation matrix
corr = df.corr()

# Plot heatmap
plt.figure(figsize=(4, 3))
sns.heatmap(
    corr,
    annot=True,       # show correlation values
    cmap="coolwarm",  # diverging colormap
    vmin=-1, vmax=1,  # fixed scale
    square=True,
    lw=1,
    cbar_kws={"shrink": 0.8}
)
plt.title("Node Attribute Correlation", pad=12)
plt.tight_layout()
plt.show()
```

# Hypergraph

```python
hyperedges = {
    1: {1, 2, 6, 7},        # cluster around node 1–2–6–7
    2: {2, 3, 4, 8},        # core connection 2–3–4–8
    3: {2, 9, 11, 12},      # branch from 2 to 9–11–12
    4: {3, 5, 10, 15},      # chain from 3 to 15
    5: {8, 13, 14},         # branch from 8 to 14
}  

# Create the HyperNetX hypergraph
H = hnx.Hypergraph(hyperedges)

plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.figsize'] = 4, 4

hnx.draw(
    H, 
    pos=G.pos, 
    fill_edges=True,
    node_label_alpha=0.0, 
    edge_label_alpha=0.0,
    with_edge_labels=False,
)

plt.gcf().set_facecolor('none')
```

```python
plt.rcParams['figure.dpi'] = 200
plt.rcParams['figure.figsize'] = 5, 4

hnx.draw_incidence_upset(H)
```

# Hypergraph Centrality

```python
# Compute centralities
B = H.incidence_dataframe()
node_ev, edge_ev = central.hevc(B, function='linear')
print(edge_ev)

# Normalize edge_ev for colormap
vmin, vmax = np.min(edge_ev), np.max(edge_ev)
norm = colors.Normalize(vmin=vmin, vmax=vmax)
cmap = cm.get_cmap("coolwarm")

# Create mapping from hyperedge IDs to color
edge_color_map = {
    e: cmap(norm(ev))
    for e, ev in zip(H.edges, edge_ev)
}

# Plot
plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.figsize'] = (4, 4)

# Create figure and axes
fig, ax = plt.subplots()

hnx.draw(
    H,
    pos=G.pos,
    fill_edges=True,
    node_label_alpha=0.0,
    edge_label_alpha=0.0,
    with_edge_labels=False,
    edges_kwargs={"color": [edge_color_map[e] for e in H.edges]}
)

# Optional colorbar
sm = cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
plt.colorbar(sm, ax=ax, shrink=0.5, label="Edge HEVC")

plt.gcf().set_facecolor('none')
plt.show()
```

# Hypergraph Ricci Curvature

```python
# Compute HLRC curvature
blist = ut.incidence_to_list_of_list(B)
hlrc = curv.compute_hlrc(blist)

# Normalize HLRC for colormap
vmin, vmax = np.nanmin(hlrc), np.nanmax(hlrc)
norm = colors.Normalize(vmin=vmin, vmax=vmax)
cmap = cm.get_cmap("coolwarm")

# Map hyperedges to colors (handle NaN values if present)
edge_color_map = {
    e: cmap(norm(val)) if val is not None and not np.isnan(val) else (0.8, 0.8, 0.8, 1.0)
    for e, val in zip(H.edges, hlrc)
}

# Plot
plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.figsize'] = (4, 4)

fig, ax = plt.subplots()

hnx.draw(
    H,
    pos=G.pos,
    fill_edges=True,
    node_label_alpha=0.0,
    edge_label_alpha=0.0,
    with_edge_labels=False,
    edges_kwargs={"color": [edge_color_map[e] for e in H.edges]},
    ax=ax
)

# Colorbar
sm = cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
fig.colorbar(sm, ax=ax, shrink=0.5, label="Edge HLRC")

plt.gcf().set_facecolor('none')
plt.show()
```
