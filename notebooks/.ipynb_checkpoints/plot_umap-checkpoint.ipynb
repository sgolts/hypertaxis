```python
# load ann data
resolution = 1000000
population_path = f"/scratch/indikar_root/indikar1/shared_data/higher_order/lightweight/population_mESC_{resolution}_anndata.h5ad"

start_time = time.time()  # Record the start time
adata = sc.read_h5ad(population_path)
end_time = time.time()  # Record the end time
print(f"Time taken to read the file: {end_time - start_time:.2f} seconds")
sc.logging.print_memory_usage()

# sort the object by global bin id
sorted_indices = adata.obs['bin'].argsort()
adata = adata[sorted_indices, :] 

adata
```

```python
def find_outliers(df_column):
  """
  Identifies outliers in a pandas DataFrame column using the IQR method.

  Args:
    df_column: A pandas Series representing the column to analyze.

  Returns:
    A boolean mask with True for outliers and False otherwise.
  """
  Q1 = df_column.quantile(0.15)
  Q3 = df_column.quantile(0.85)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  return (df_column < lower_bound) | (df_column > upper_bound)

adata.obs['degree_outlier'] = find_outliers(adata.obs['degree'])

print(adata.obs['degree_outlier'].value_counts())

# remove outliers
remove_bins = adata.obs[adata.obs['degree_outlier']].index.to_list()
print(f"Removing top {len(remove_bins)} outlier loci: ")
print(remove_bins)

adata = adata[~adata.obs_names.isin(remove_bins), :].copy()

print('done!')
```

# Define the core

```python
order_threshold = 3
score_column = 'global_hge_logexp_RNA_weighted'
core_threshold_quantile = 0.75
threshold = np.quantile(adata.obs[score_column], core_threshold_quantile)

print(f"Using score column: {score_column}")
print(f"Core threshold quantile: {core_threshold_quantile}")
print(f"Calculated threshold: {threshold:.3f}") 

core_nodes = adata.obs[adata.obs[score_column] > threshold].index.to_list()

print(f"Detected {(len(core_nodes))} of {len(adata.obs)} total nodes ({100*((len(core_nodes)) / len(adata.obs)):.1f}%)")

# extract the core from population
core = adata[core_nodes, :].copy()
core = core[:, core.X.sum(axis=0) > order_threshold].copy()

# add the clique-expanded representation
matrix.expand_and_normalize_anndata(core)

core
```

```python
plot_order = 3
detection_threshold = 12
detected_bins = qdata.obs_names[qdata.to_df().sum(axis=1) > detection_threshold]
print(len(detected_bins))

""" GET THE CHROMOSOME INFORMATION"""
chroms = list(set(detected_bins.str.split(":").str[0]))
print(chroms)

pdf = df.copy()
pdf['chrom'] = pdf['bin_name_y'].str.split(":").str[0]
pdf = pdf[pdf['chrom'].isin(chroms)]
pdf = pdf.sort_values(by='bin')

pdf = pdf.groupby('bin_name_y', sort=False).agg(
    UMAP_1=('UMAP 1', 'mean'),
    UMAP_2=('UMAP 2', 'mean'),
    n_genes=('n_genes', 'sum'),
    ATACSeq_1=('ATACSeq_1', 'mean'),
    CTCF=('CTCF', 'mean'),
    H3K27ac=('H3K27ac', 'mean'),
    RNA_5=('RNA_5', 'mean'),
    core_score=('global_hge_logexp_RNA_weighted', 'mean'),
    
    is_core=('is_core', 'max'),
).reset_index()


""" BUILD THE EDGELIST """
edge_list =  pd.DataFrame({
    'node1' : pdf['bin_name_y'].values,
    'node2': pdf['bin_name_y'].shift(-1).values
}).dropna()

edge_list['node_1_chr'] = edge_list['node1'].str.split(":").str[0]
edge_list['node_2_chr'] = edge_list['node2'].str.split(":").str[0]

# drop inter-chromosomal edges
edge_list = edge_list[edge_list['node_1_chr'] == edge_list['node_2_chr']]

G = nx.from_pandas_edgelist(
    edge_list,
    source='node1',
    target='node2',
)

"""ADD NODE INFORMAION"""

node_attributes = pdf.copy()
node_attributes = node_attributes.set_index('bin_name_y')
node_attributes = node_attributes.to_dict(orient='index')
nx.set_node_attributes(G, node_attributes)
print(G)

""" BUILD THE LAYOUT BASED ON UMAP """

rows = zip(pdf['bin_name_y'], pdf['UMAP_1'].values, pdf['UMAP_2'].values)
init_positions = {bin_name: (umap1, umap2) for bin_name, umap1, umap2 in rows}

G.pos = nx.spring_layout(
    G,
    pos=init_positions,
    k=0.75,
    seed=32,
)

""" DEFINE COLORS """
chrom_colors = plt2.get_n_colors(len(chroms)+2, cmap_name=chrom_cmap)
chrom_colors = dict(zip(chroms, chrom_colors[1:-1]))
edge_colors = edge_list['node_1_chr'].map(chrom_colors).values
node_colors = ['none' if G.nodes[node]['is_core'] else 'none' for node in G.nodes]

""" Create the legend """
fig_legend, ax_legend = plt.subplots(figsize=(2, 4))  
legend_patches = [mpatches.Patch(color=color, label=label)
           for label, color in chrom_colors.items()]

ax_legend.legend(handles=legend_patches, loc='center', title='chromosome')
ax_legend.axis('off')
plt.show()

""" PLOT """
plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.figsize'] = 10, 8

nx.draw_networkx_nodes(
    G, 
    G.pos,
    node_size=45,
    node_color=node_colors,
    edgecolors='none', 
    linewidths=1.0,
)

nx.draw_networkx_edges(
    G, 
    G.pos,
    width=10,
    edge_color=edge_colors,
    node_size=0,
)

""" PLOT HYPEREDGES """
H = qdata.to_df().T
H = H[detected_bins]
H = H[H.sum(axis=1) == 3]
print(f"{H.shape=}")

def get_non_zero_cols(row):
  return sorted(list(H.columns[row.astype(bool)]))

non_zero_cols = H.apply(get_non_zero_cols, axis=1)

ax = plt.gca()
for hyperedge in non_zero_cols.value_counts().index:
    node_chrom = set([node.split(":")[0] for node in hyperedge])
    if len(node_chrom) < 2:
        continue

    polygon_coords = [G.pos[node] for node in hyperedge]
    
    poly = patches.Polygon(
        polygon_coords, 
        closed=True, 
        alpha=0.25, 
        facecolor='grey',
        edgecolor='k',
        linewidth=2,
        zorder=30,
    )  
    ax.add_patch(poly)

    # Add points on vertices
    for coord in polygon_coords:
        ax.plot(
            coord[0], 
            coord[1],
            marker='o', 
            color='r',
            markersize=7,
            zorder=31,
            mec='k',
        ) 

    # # Annotate the points
    # for node in hyperedge:
    #     x, y = G.pos[node]
    #     ax.annotate(node, (x, y), textcoords="offset points", xytext=(5,5), ha='center', fontsize=8)
```
