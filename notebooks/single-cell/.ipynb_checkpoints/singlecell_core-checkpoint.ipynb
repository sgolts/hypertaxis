```python
fpath = "/scratch/indikar_root/indikar1/shared_data/higher_order/transcription_clusters/singlecell_core_incidence_1000000_protien_coding_only.pkl"
df =  pd.read_pickle(fpath)
print(f"{df.shape=}")
df.head()
```

# most common genes

```python
pdf = df.T.copy()
pdf = pdf.astype(bool).astype(int)
pdf['cell_id'] = pdf.index.str[:5]
pdf = pdf.groupby('cell_id').max()
print(f"{pdf.shape=}")
pdf.head()
```

```python
gdf = pdf.sum(axis=0).reset_index()
gdf.columns = ['gene_name', 'cell_count']
gdf = gdf.sort_values(by='cell_count', ascending=False)
gdf['percent'] = 100*(gdf['cell_count'] / pdf.shape[0])
print(gdf.head(10).to_string(index=False))
```
