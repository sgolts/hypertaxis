# Load the duplicate reports



```python
resolution = 1000000
dpath = f"/scratch/indikar_root/indikar1/shared_data/higher_order/duplicates/singlecell_mESC_{resolution}_chr*.parquet"
file_list = glob.glob(dpath)
print(f"n files: {len(file_list)}")

df = []

for fpath in file_list:
    chrom = os.path.basename(fpath).split("_")[3].split(".")[0]
    print(f"{chrom}...")
    tmp = pd.read_parquet(fpath)
    tmp['chrom'] = chrom
    df.append(tmp)

df = pd.concat(df)
df = df.sort_values(by=['read_id', 'mapping_quality'], ascending=[True, False])
df = df.drop_duplicates(subset=['read_id', 'cell_id'])

print(df['unique'].value_counts())
print(f"{df.shape=}")
df.head()
```

```python
outpath = f"/scratch/indikar_root/indikar1/shared_data/higher_order/duplicate_table/singlecell_mESC_{resolution}_dupes.parquet"

df.to_parquet(outpath)
df.head()
```
