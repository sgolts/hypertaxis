```python
outpath = "/scratch/indikar_root/indikar1/shared_data/higher_order/expression_table/rna_table.parquet"

""" READ expression information """
fpath = "/nfs/turbo/umms-indikar/shared/projects/poreC/data/kauffmanRNA/4DNFIC269AEU.tsv"
tdf  = pd.read_csv(fpath, sep='\t')
tdf = tdf[tdf['gene_id'].str.startswith("ENS")]
tdf['ens_gene_id'] = tdf['gene_id'].str.split(".").str[0]
print(f"(raw) {tdf.shape=}")


""" ADD gene names """
fpath = "/scratch/indikar_root/indikar1/shared_data/higher_order/reference/gene_table.parquet"
gdf = pd.read_parquet(fpath, columns=['gene_name', 'gene_id'])
print(f"{gdf.shape=}")
gene_map = dict(zip(gdf['gene_id'].values, gdf['gene_name'].values))
tdf['gene_name'] = tdf['ens_gene_id'].map(gene_map)

# drop unmappable
tdf = tdf[tdf['gene_name'].notna()]
print(f"(filtered) {tdf.shape=}")

tdf.to_parquet(outpath, index=False)

tdf.sample(5)
```
