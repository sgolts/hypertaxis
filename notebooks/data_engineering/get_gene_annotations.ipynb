```python
import pandas as pd
```

# PangloaDB

```python
pangloa_path = "/nfs/turbo/umms-indikar/shared/projects/spatial_transcriptomics/data/panglaodb/pandb.tsv.gz"
pdf = pd.read_csv(pangloa_path, sep="\t")
pdf = pdf[pdf['species'].notna()]

pdf = pdf[pdf['species'].str.contains('Mm')]

sorted(pdf['cell type'].unique())



cell_types = [
    'Embryonic stem cells',
    'Pluripotent stem cells',
]

pdf = pdf[pdf['cell type'].isin(cell_types)].reset_index(drop=True)

keep_columns = [
    'official gene symbol', 
    'cell type',
    'ubiquitousness index',
    'sensitivity_mouse',
    'specificity_mouse',
]

pdf = pdf[keep_columns]
pdf.columns = [
    'gene_name', 
    'cell_type',
    'ubiquitousness',
    'sensitivity_mouse',
    'specificity_mouse',
]

pdf.to_csv("../../resources/mESC_pangloadb.csv", index=False)

pdf.head()
```
